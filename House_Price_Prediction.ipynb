```python
import numpy as np          # arrays
import pandas as pd          # dataframe
import matplotlib.pyplot as plt       # visualizing (plots & graphs)
import seaborn as sns         # plotting
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics      # evaluation
```

Importing Boston House Price Dataset

```python
house_price_dataset = sklearn.datasets.load_boston()
```

```python
print(house_price_dataset)
```

```python
# loading the dataset to pandas dataframe
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
```

```python
# Add the target column (price) to the dataframe
house_price_dataframe['Price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
# checking the number of rows and columns
house_price_dataframe.shape
```

```python
# Statistical measures
house_price_dataframe.describe()
```

```python
#checking for missing values
house_price_dataframe.isnull().sum()
```

Understanding the correlation between various features in the dataset

1. Positive correlation
2. Negative correlation

```python
correlation = house_price_dataframe.corr()
```

```python
#constructing a heatmap to understand the correlatioin
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Reds')
```

Splitting the data and target

```python
X = house_price_dataframe.drop(['Price'], axis=1)
Y = house_price_dataframe['Price']
```

```python
print(X, Y)
```

Splitting the data into Training and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training

XGBoost Regressor

```python
#loading the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
#accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error: ", score_1)
print("Mean Absolute Error: ", score_2)
```

Visualizing the Actual prices and Predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices VS Prediceted Prices")
plt.show()
```

Prediction on testing data

```python
#accuracy for prediction in testing data
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
# R squared error
score_11 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_22 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error: ", score_11)
print("Mean Absolute Error: ", score_22)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices VS Prediceted Prices")
plt.show()
```
